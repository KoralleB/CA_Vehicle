### Choosing best resampling technique for logistic regression

```python
fpr_log, tpr_log = np.load('../output_files/fpr_log.npy'),np.load('../output_files/tpr_log.npy')
roc_auc_log = np.load('../output_files/roc_auc_log.npy')

fpr_log_over, tpr_log_over = np.load('../output_files/fpr_log_over.npy'),np.load('../output_files/tpr_log_over.npy')
roc_auc_log_over = np.load('../output_files/roc_auc_log_over.npy')

fpr_log_smote, tpr_log_smote = np.load('../output_files/fpr_log_smote.npy'),np.load('../output_files/tpr_log_smote.npy')
roc_auc_log_smote = np.load('../output_files/roc_auc_log_smote.npy')

label = ['log','log_over','log_smote']
fpr = [fpr_log,fpr_log_over,fpr_log_smote]
tpr = [tpr_log,tpr_log_over,tpr_log_smote]
roc_auc = [roc_auc_log,roc_auc_log_over,roc_auc_log_smote]
roc_plot(label,fpr,tpr,roc_auc)
```

```python
precision_log, recall_log = np.load('../output_files/precision_log.npy'),np.load('../output_files/recall_log.npy')
precision_log_over, recall_log_over = np.load('../output_files/precision_log_over.npy'),np.load('../output_files/recall_log_over.npy')
precision_log_smote, recall_log_smote = np.load('../output_files/precision_log_smote.npy'),np.load('../output_files/recall_log_smote.npy')

label = ['log','log_over','log_smote']
recall = [recall_log,recall_log_over,recall_log_smote]
precision = [precision_log,precision_log_over,precision_log_smote]
pr_plot(label,recall,precision)
```

```python
# confusion matrix
cm_log = np.load('../output_files/cm_log.npy')
cm_log_over = np.load('../output_files/cm_log_over.npy')
cm_log_smote = np.load('../output_files/cm_log_smote.npy')

label = ['log','log_over','log_smote']
cm = [cm_log, cm_log_over, cm_log_smote]
conf_mat(label, cm)
```

```python
# Accuracy
acc_log = np.load('../output_files/acc_log.npy')
acc_log_over = np.load('../output_files/acc_log_over.npy')
acc_log_smote = np.load('../output_files/acc_log_smote.npy')

label = ['log','log_over','log_smote']
acc = [acc_log, acc_log_over, acc_log_smote]
acc_print(label,acc)
```

The best resampling technique is SMOTE, since it has the best PR curve and good CM.
