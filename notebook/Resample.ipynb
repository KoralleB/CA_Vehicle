```python
precision_log, recall_log = np.load('../output_files/precision_log.npy'),np.load('../output_files/recall_log.npy')
precision_log_over, recall_log_over = np.load('../output_files/precision_log_over.npy'),np.load('../output_files/recall_log_over.npy')
precision_log_smote, recall_log_smote = np.load('../output_files/precision_log_smote.npy'),np.load('../output_files/recall_log_smote.npy')

label = ['log','log_over','log_smote']
recall = [recall_log,recall_log_over,recall_log_smote]
precision = [precision_log,precision_log_over,precision_log_smote]
pr_plot(label,recall,precision)
```

```python
# Accuracy
acc_log = np.load('../output_files/acc_log.npy')
acc_log_over = np.load('../output_files/acc_log_over.npy')
acc_log_smote = np.load('../output_files/acc_log_smote.npy')

label = ['log','log_over','log_smote']
acc = [acc_log, acc_log_over, acc_log_smote]
acc_print(label,acc)
```
